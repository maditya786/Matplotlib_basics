```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,4*np.pi,100)
# print(x)
# print([i for i in sin_y])
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,"--",label="Sine")
plt.plot(x,cos_y,"*",label="cos")

plt.legend(loc = 'upper right')

plt.title("Sine Wave")
plt.show()
```
